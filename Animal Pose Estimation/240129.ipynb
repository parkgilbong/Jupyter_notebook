```python
df_mouse = pd.read_csv("long format.csv", 
                       index_col = 0)   
# df_mouse.columns = ['M1SnoutX', 'M1SnoutY', 'M1SnoutLH', 'M1EarRX', 'M1EarRY', 'M1EarRLH', 'M1EarLX', 'M1EarLY', 'M1EarLLH', 'M1CentroidX', 'M1CentroidY', 'M1CentroidLH', 'M1TbX', 'M1TbY', 'M1TbLH', 'M2SnoutX', 'M2SnoutY', 'M2SnoutLH', 'M2EarRX', 'M2EarRY', 'M2EarRLH', 'M2EarLX', 'M2EarLY', 'M2EarLLH', 'M2CentroidX', 'M2CentroidY', 'M2CentroidLH', 'M2TbX', 'M2TbY', 'M2TbLH', 'single1X', 'single1Y', 'single1LH', 'single2X', 'single2Y', 'single2LH']
df_mouse = df_mouse.replace("", np.nan)
df_mouse.head()
```

```python
XYtrace = sns.lineplot(data=df_mouse, x='centroid_x', y='centroid_y', hue='ID', sort = False)
XYtrace.set(xlim = (200, 1400), ylim = (200, 1400), xlabel = "", ylabel = "")
XYtrace.invert_yaxis()
sns.set(rc={'figure.figsize':(5,5)})
sns.despine()
```
