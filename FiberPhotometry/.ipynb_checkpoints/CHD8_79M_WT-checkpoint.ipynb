Streams 데이터에 접근

```python
# isosbestic control signal 
FPdata.streams['_405A']
```

```python
# export isosbestic and GCaMP signals
nd_control = FPdata.streams['_405A'].data
nd_signal = FPdata.streams['_465A'].data
nd_control, nd_signal
```

```python
#control sample 수 확인
num_sample = len(nd_control)
num_sample
```

```python
#control sampling rate 확인
FPdata.streams['_405A'].fs
```

```python
# X-axis 생성
wav_time = np.linspace(1, num_sample, num_sample)/FPdata.streams['_405A'].fs
wav_time
```

```python
fig_405A = plt.subplots(figsize=(10,6))
plt.plot(wav_time, nd_control)
plt.show()
```

```python
fig_465A = plt.subplots(figsize=(10,6))
plt.plot(wav_time, nd_signal)
plt.show()
```

```python
# plot signals
fig,ax1=plt.subplots()  # create a plot to allow for dual y-axes plotting
plot1=ax1.plot(wav_time, nd_control, 'grey', label='405') #plot 405 on left y-axis
ax2=plt.twinx()# create a right y-axis, sharing x-axis on the same plot
plot2=ax2.plot(wav_time, nd_signal, 'g', label='GCaMP6f') # plot GCaMP on right y-axis

# Plot rewards times as ticks.
camera_ticks = ax1.plot(FPdata.epocs.Cam1.onset, np.full(np.size(FPdata.epocs.Cam1.onset), 68), label='camera ticks', color='w', marker="|", mec='k')


ax1.set_ylim(35, 70)
ax1.set_xlim(0, 5) # 초단위로 입력하여 x-axis zoom in 가능 
ax2.set_ylim(35, 70)
ax1.set_xlabel('Time (seconds)')
ax1.set_ylabel('405 Signal', color='grey')
ax2.set_ylabel('GCaMP6f Signal', color='g')
ax1.set_title('Raw signals')

lines = plot1 + plot2 +camera_ticks #line handle for legend
labels = [l.get_label() for l in lines]  #get legend labels
legend = ax1.legend(lines, labels, loc='lower right', bbox_to_anchor=(1, 1)) #add legend
```
